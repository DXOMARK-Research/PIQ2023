```python
import os
import pandas as pd
```

```python
scorePath = r"./Scores/Scores_Exposure.csv"

scoreDf = pd.read_csv(scorePath)
```

```python
scoreDf
```

### Scene Split (Generalization Split)

```python
sceneSplitPath = r"./Test split/Scene Split.csv"
sceneSplitDf = pd.read_csv(sceneSplitPath)
```

```python
sceneSplitDf
```

```python
trainDf = pd.merge(scoreDf, sceneSplitDf[sceneSplitDf['SPLIT']=='Train'].drop('NB IMAGES', axis=1), on='SCENE', how='inner')
testDf = pd.merge(scoreDf, sceneSplitDf[sceneSplitDf['SPLIT']=='Test'].drop('NB IMAGES', axis=1), on='SCENE', how='inner')

trainDf.to_csv(rf'./Test split/Scene Split/SceneSplit_Train_{os.path.basename(scorePath)}', index=False)
testDf.to_csv(rf'./Test split/Scene Split/SceneSplit_Test_{os.path.basename(scorePath)}', index=False)
```
